### 正态分布检验
##### 可视化：直方图 Q-Q图
##### 正态性检验：Shapiro-Wilk检验 Kolmogorov-Smirnov检验 Anderson-Darling检验
##### 偏度和峰度检验

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 设置随机种子，以保证结果可重复
np.random.seed(823)

# 生成数据
normal_data = np.random.normal(loc=0, scale=1, size=1000)  # 正态分布数据
exponential_data = np.random.exponential(scale=1, size=1000)  # 指数分布数据

# 可视化：直方图和Q-Q图
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# 正态分布数据
axs[0, 0].hist(normal_data, bins=30, density=True, alpha=0.6, color='b')
axs[0, 0].set_title("Normal Distribution - Histogram")
stats.probplot(normal_data, dist="norm", plot=axs[0, 1])
axs[0, 1].set_title("Normal Distribution - Q-Q Plot")

# 指数分布数据
axs[1, 0].hist(exponential_data, bins=30, density=True, alpha=0.6, color='r')
axs[1, 0].set_title("Exponential Distribution - Histogram")
stats.probplot(exponential_data, dist="norm", plot=axs[1, 1])
axs[1, 1].set_title("Exponential Distribution - Q-Q Plot")

plt.tight_layout()
plt.show()

# 正态性检验

# 1. Shapiro-Wilk Test
normal_shapiro_stat, normal_shapiro_p = stats.shapiro(normal_data)
exponential_shapiro_stat, exponential_shapiro_p = stats.shapiro(exponential_data)

# 2. Kolmogorov-Smirnov Test
normal_ks_stat, normal_ks_p = stats.kstest(normal_data, 'norm')
exponential_ks_stat, exponential_ks_p = stats.kstest(exponential_data, 'norm')

# 3. D'Agostino's Test
normal_dagostino_stat, normal_dagostino_p = stats.normaltest(normal_data)
exponential_dagostino_stat, exponential_dagostino_p = stats.normaltest(exponential_data)

# 输出结果
print(f"Shapiro-Wilk Test for Normal Data: p-value = {normal_shapiro_p}")
print(f"Shapiro-Wilk Test for Exponential Data: p-value = {exponential_shapiro_p}")
print(f"Kolmogorov-Smirnov Test for Normal Data: p-value = {normal_ks_p}")
print(f"Kolmogorov-Smirnov Test for Exponential Data: p-value = {exponential_ks_p}")
print(f"D'Agostino Test for Normal Data: p-value = {normal_dagostino_p}")
print(f"D'Agostino Test for Exponential Data: p-value = {exponential_dagostino_p}")
```
