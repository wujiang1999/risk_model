```python
import numpy as np
import pandas as pd

# 假设 model_1 和 model_2 是已训练的模型，且它们的预测结果为概率值
# 以下为示例的预测结果
pred_1 = np.array([0.8, 0.3, 0.6, 0.9, 0.2])  # model_1 的预测概率
pred_2 = np.array([0.7, 0.4, 0.5, 0.8, 0.3])  # model_2 的预测概率

# 权重设置（w1 和 w2 的和应为 1）
w1 = 0.6  # model_1 的权重
w2 = 0.4  # model_2 的权重

# Step 1: 对每个模型的预测结果进行排序，获得排序后的分位点（排名）
# 获取 model_1 和 model_2 在排序后的分位点
rank_1 = pd.Series(pred_1).rank(ascending=False).values  # 从大到小的排名
rank_2 = pd.Series(pred_2).rank(ascending=False).values  # 从大到小的排名

# Step 2: 计算加权后的分位点
weighted_rank = w1 * rank_1 + w2 * rank_2

# Step 3: 根据加权后的分位点排序得到最终的预测结果
# 你可以选择根据加权后的排名进行某些操作，例如选择排名最高的项
final_prediction = pd.Series(weighted_rank).rank(ascending=True).values  # 从小到大的排名

# 输出最终的加权分位点
print("加权后的分位点：", final_prediction)
```
